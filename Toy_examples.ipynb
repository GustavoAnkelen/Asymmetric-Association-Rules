```python
import os
os.chdir("C:\\Users\gusta\JupyterLab\Laboratorio de Modelacion 2\Data")

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from itertools import combinations

import csv

transactions = pd.read_csv("data_assorules.csv")
category = pd.read_csv("cat_nombres.csv")
department = pd.read_csv("data_assorules_hat.csv")
categorias = sorted(transactions['Category'].unique())
```

```python
soportes  = pd.read_csv("soporte_conjunto.csv").drop('Unnamed: 0',axis=1).set_axis(categorias).rename(columns=int)

lift = pd.read_csv("lift.csv").drop('Unnamed: 0',axis=1).set_axis(categorias).rename(columns=int)
confidence =  pd.read_csv("confidence.csv").drop('Unnamed: 0',axis=1).set_axis(categorias).rename(columns=int)
AddedValue = pd.read_csv("added_value.csv").drop('Unnamed: 0',axis=1).set_axis(categorias).rename(columns=int)
conviction = pd.read_csv("conviction.csv").drop('Unnamed: 0',axis=1).set_axis(categorias).rename(columns=int)
collective = pd.read_csv("collective_strength.csv").drop('Unnamed: 0',axis=1).set_axis(categorias).rename(columns=int)
```

```python
toy_supp = soportes[soportes.columns.intersection(category['Original'])].loc[soportes.index.intersection(category['Original'])]
toy_lift = lift[lift.columns.intersection(category['Original'])].loc[lift.index.intersection(category['Original'])]
toy_confidence =  confidence[confidence.columns.intersection(category['Original'])].loc[confidence.index.intersection(category['Original'])]
toy_AddedValue =  AddedValue[AddedValue.columns.intersection(category['Original'])].loc[AddedValue.index.intersection(category['Original'])]
toy_conviction =  conviction[conviction.columns.intersection(category['Original'])].loc[conviction.index.intersection(category['Original'])]
toy_collective =  collective[collective.columns.intersection(category['Original'])].loc[collective.index.intersection(category['Original'])]
```

```python
# Reemplazamos directamente, pues las categorias vienen ordenadas
[toy_supp.columns,toy_supp.index] = [category['Nombre'],category['Nombre']] 
[toy_lift.columns,toy_lift.index] = [category['Nombre'],category['Nombre']] 
[toy_confidence.columns,toy_confidence.index] = [category['Nombre'],category['Nombre']]
[toy_AddedValue.columns,toy_AddedValue.index] = [category['Nombre'],category['Nombre']]
[toy_conviction.columns, toy_conviction.index] = [category['Nombre'],category['Nombre']]
[toy_collective.columns, toy_collective.index] = [category['Nombre'],category['Nombre']]

toy_supp.to_csv('ToySupp.csv', encoding='utf_8')
toy_lift.to_csv('ToyLift.csv', encoding='utf_8')
toy_AddedValue.to_csv('ToyConfidence.csv', encoding='utf_8')
toy_confidence.to_csv('ToyConfidence.csv', encoding='utf_8')
toy_conviction.to_csv('ToyConviction.csv', encoding='utf_8')
toy_collective.to_csv('ToyCollectiveStrenght.csv', encoding= 'utf_8')
```
